```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
raw_document = TextLoader("tagged_description.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_document)
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(documents, embedding=OpenAIEmbeddings())
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split(":")[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)  # Retrieve top 50 similar books

    books_list = []

    for i in range(len(recs)):
        isbn_candidate = recs[i].page_content.strip('"').strip(":").split()[0]

        # Remove non-numeric characters (e.g., colons, dashes, etc.)
        clean_isbn = ''.join(filter(str.isdigit, isbn_candidate))

        # Convert safely if ISBN is valid
        if clean_isbn.isdigit():
            books_list.append(clean_isbn)  # Keep as string to match dataframe dtype
        else:
            print(f"Skipping invalid ISBN: {isbn_candidate}")  # Debugging message

    # Filter books by ISBN list
    filtered_books = books[books["isbn13"].astype(str).isin(books_list)]

    return filtered_books.head(top_k)  # Apply head() correctly
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
