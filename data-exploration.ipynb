```python
from statistics import correlation

import kagglehub
from seaborn import heatmap

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Column")
plt.ylabel("Missing values")

plt.show()
```

```python
import numpy as np

books["missing_descriptions"] = np.where(books["description"].isna(), 1, 0)  #missing - 1, not missing - 0
books['age_of_book'] = 2025 - books['published_year']
```

```python
 books
```

```python
columns_of_interest = ['num_pages', 'age_of_book', 'missing_descriptions', 'average_rating']

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap="coolwarm", cbar_kws={'label': 'Spearman Correlation'})
heatmap.set_title("Spearman Correlation Heatmap")
plt.show()
```

```python
books[(books["description"].isna() | books["average_rating"].isna() | books["num_pages"].isna() | books["published_year"].isna())]
```

```python
#since it is only 303 rows, <5% of the given data, we can remove them
book_missing = books[(~books["description"].isna() | ~books["average_rating"].isna() | ~books["num_pages"].isna() | ~books["published_year"].isna())]
book_missing
```

```python
#sorting categories in descending order based on their category count to find the most common/popular categories
book_missing['categories'].value_counts().reset_index().sort_values(by='count', ascending=False)
```

```python
book_missing
```

```python
#length of description to get the count of number of words to know whether the description is good or not
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
#finding useless description - too short description is not good enough
book_missing.loc[book_missing["words_in_description"].between(5,14), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(15,24), "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words  #5230 rows is still good enough for a recommender
```

```python
#where clause states - if subtitle is empty use title instead but if not empty, combine them
#astype converted to string because empty values are considered as float in python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join, axis=1))
)
#the warning is coming because the changes are done directly into the main dataframe and not in a copy
```

```python
book_missing_25_words
```

```python
#isbn13 value is attached to the beginning of the description as it will help in vector generation
book_missing_25_words["tagged_description"] = book_missing_25_words[['isbn13', 'description']].astype(str).agg(': '.join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(columns=["subtitle", "missing_descriptions", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
